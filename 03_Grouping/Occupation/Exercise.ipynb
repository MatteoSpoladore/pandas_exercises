### Step 1. Import the necessary libraries


```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).


### Step 3. Assign it to a variable called users.


```python
users = pd.read_table(
    "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user",
    sep="|",
    index_col="user_id",
)
users.head()
```

### Step 4. Discover what is the mean age per occupation


```python
users.groupby(by="occupation")[["age"]].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least


```python
users[users.gender == "M"].groupby(by="occupation")[["age"]].mean().sort_values(
    by="age", ascending=False
)

users["gender_n"] = users["gender"].apply(lambda x: 1 if x == "M" else 0)
a = round(
    users.groupby("occupation").gender_n.sum() / users.occupation.value_counts() * 100,
    2,
)
a.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages


```python
users.groupby(by="occupation")[["age"]].agg(["mean", "max", "min"])
```

### Step 7. For each combination of occupation and gender, calculate the mean age


```python
users.groupby(by=["occupation", "gender"])["age"].agg(["mean", "max", "min"])
```

### Step 8. For each occupation present the percentage of women and men


```python
# size conta le righe
# unstack trasforma il livello più interno dell'indice (in questo caso gender) in colonne
# fill value = 0 serve per aggiungere 0 se non c'è un gendere in un a determinata occupation
counts = users.groupby(["occupation", "gender"]).size().unstack(fill_value=0)

counts["total"] = counts.F + counts.M
counts["%F"] = round(counts.F / counts.total * 100, 2)
counts["%M"] = 100.00 - counts["%F"]

counts
```
