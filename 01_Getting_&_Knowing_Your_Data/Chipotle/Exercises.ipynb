### Step 3. Assign it to a variable called chipo.

### Step 9. Which was the most-ordered item?


```python
chipo.groupby(by="item_name")["item_name"].count().sort_values(ascending=False)[:1]
```

### Step 13. Turn the item price into a float


#### Step 13.a. Check the item price type


```python
display(chipo.item_price.dtype)
display(chipo.dtypes)
display(chipo.head())
```

#### Step 13.b. Create a lambda function and change the type of item price


```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x.replace("$", "")))
```

#### Step 13.c. Check the item price type


```python
chipo.item_price.dtype
```

### Step 14. How much was the revenue for the period in the dataset?


```python
sum(chipo.quantity * chipo.item_price)
```

### Step 15. How many orders were made in the period?


```python
print(len(chipo.order_id.unique()))
print(chipo.order_id.value_counts())
print(chipo.order_id.value_counts().count())
```

### Step 16. What is the average revenue amount per order?


```python
# Solution 1

chipo["revenue"] = chipo.item_price * chipo.quantity
display(chipo.head())
display(chipo.groupby(by="order_id")[["revenue"]].sum())
chipo.groupby(by="order_id")["revenue"].sum().mean()
```
